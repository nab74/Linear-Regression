# Linear Regression using Standardized Data

**Loading the Data Set:**

```python
from re import M
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np


marks = pd.read_csv("https://raw.githubusercontent.com/tofighi/MachineLearning/master/datasets/student_marks.csv")
#print(marks)
m=-0.5
b=0
x1=marks['Midterm mark'].values
y1=marks['Final mark'].values
```

**Midterm and Final Mean Calculations:**

```python
#Your code
mid_mean=x1.mean()
final_mean=y1.mean()
```

**Midterm and Final Standard Deviation Calculations:**

```python
#Your code
mid_std=x1.std()
final_std=y1.std()
```

**Standardization:**

```python
#Your code
def mid_standardize():
  mid_marks=(x1-mid_mean)/mid_std
  return mid_marks
def fin_standardize():
  fin_marks=(y1-final_mean)/final_std
  return fin_marks

X_std=mid_standardize()
Y_std=fin_standardize()
```

**Graph #1 Initial Values:**

```python
#student marks
plt.scatter(X_std,Y_std,color='blue', label='Data Points')
plt.title("Standardized Initial Student Marks")
plt.xlabel("midterm")
plt.ylabel("final")

#initial line
x=np.linspace(np.min(X_std), np.max(Y_std),100)
y=m*x+b
plt.plot(x,y,'-r', label='Initial regression line')
plt.legend()
```

**Error (Cost Function) Initialization:**

```python
#Your code
def cost_function(y_true,x_pred):
  error=(((y_true-(m*x_pred+b))**2).sum())/100
  return error
```

**Gradient Partial Derivatives:**

```python
def derivatives(X, y, m, b):
#partial derivative w.r.t m
  dm=(2/100)*np.dot(X.T, (np.dot(X,m)+b))
  
 #partial derivative w.r.t b
  db=(2/100)*np.sum(np.dot(X,m)+b)
  return db,dm
```

**Updating m and b using:**

```python
def update_mb(m, b,dm,db,rate):
#update m
  m=m-rate*dm
#update b
  b=b-rate*db
  return m,b
```

**Regression line plot function:**

```python
def plot_regression_line(X, y, m, b, n):
    plt.scatter(X, y, color='blue', label='Data Points')
    plt.plot(X, np.dot(X, m) + b, color='red', label='Regression Line')
    plt.xlabel('Midterm')
    plt.ylabel('Final')
    plt.title("Student Marks for "+str(n)+" Iterations")
    plt.legend()
    plt.show()
```

**Updating m and b (100 iterations):**

```python
#Your code
iteration_list=[]
error_list=[]

for i in range(100):
  dm,db=derivatives(X_std,Y_std, m, b)
  error=cost_function(Y_std, np.dot(X_std,m)+b)

  m, b=update_mb(m,b,dm,db,0.0001)
  
  iteration_list.append(i+1)
  error_list.append(error)

print("100th: m =", m,", b =", b)
plot_regression_line(x1,y1,m,b,100)
```

**Error Plotting function**

```python
#Your code
plt.plot(iteration_list, error_list)
plt.title('Error of 100 Iteration')
plt.xlabel('Iteration')
plt.ylabel('Error')
```

**Error at each iteration after 100 iterations:**

```python
#Your code
for i in range(100):
  print("iteration:",iteration_list[i],", error:", error_list[i])  
```

**Updating m and b (2000 iterations):**

```python
# for 2000 iteration
for i in range(2000):
  dm,db=derivatives(X_std,Y_std, m, b)
  error=cost_function(Y_std, np.dot(X_std,m)+b)

  m, b=update_mb(m,b,dm,db,0.0001)
  
  iteration_list.append(i+1)
  error_list.append(error)

print("2000th: m =", m,", b =", b)
plot_regression_line(x1,y1,m,b,2000)
```

**Error for each iteration after 2000 iterations:**

```python
#errorPlot(2000,err)
plt.plot(iteration_list, error_list)
plt.title('Error of 2000 Iteration')
plt.xlabel('Iteration')
plt.ylabel('Error')
```
